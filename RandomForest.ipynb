```python
data['Gender']=data['Gender'].map({'Male': 1,'Female':0})
data['Married']=data['Married'].map({'Yes':1,'No':0})
data['Education']=data['Education'].map({'Graduate':1,'Not Graduate':0})
data['Self_Employed']=data['Self_Employed'].map({'Yes':1,'No':0})
data['Dependents'].replace('3+',3,inplace=True)
data['Property_Area']=data['Property_Area'].map({'Semiurban':1, 'Urban':0,'Rural':2})
data['Loan_Status']=data['Loan_Status'].map({'Y':1,'N':0})
```

```python
#replace null values(replace in beforehand if using label encoding)
rev_null=['Gender','Married','Dependents','Self_Employed','Credit_History','LoanAmount','Loan_Amount_Term','Property_Area']
data[rev_null]=data[rev_null].replace({np.nan:data['Gender'].mode(),
                                       np.nan:data['Married'].mode(),
                                       np.nan:data['Dependents'].mode(),
                                       np.nan:data['Self_Employed'].mode(),
                                       np.nan:data['Credit_History'].mode(),
                                       np.nan:data['LoanAmount'].mode(),
                                       np.nan:data['Loan_Amount_Term'].mode()})
```

```python
for i in rev_null:
  data[i]= data[i].fillna(data[i].mode()[0])
```

```python
data.isnull().sum()
```

```python
x=data.drop(columns=['Loan_ID','Loan_Status']).values #independent variable
y=data['Loan_Status'].values #dependent variable

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=20)
```

```python
x_train #when biggest value is passed accuracy gets affected
#high bias , low variance - underfit
```

```python
#standard scaling - 0 to 1
#min-max scaling- within range

from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
x_train = sta.fit_transform(x_train)
x_test = sta.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier #combination of 2 or more algorithm(onsemble)
from sklearn.metrics import classification_report
rfc= RandomForestClassifier(criterion = 'entropy', random_state=42)
rfc.fit(x_train,y_train)
```

```python
#evaluating training set
y_pred_train = rfc.predict(x_test)
print(classification_report(y_test, y_pred_train))
```
